### Linear Regression with multiple features

Bad Implementation

```python
ones = np.ones((X.shape[0], 1))
X = np.hstack((ones, X))
```

```python
def hypothesis(x, theta):
    
    y_ = 0.0
    n = x.shape[0]
    
    for i in range(n):
        y_ = y_ + (theta[i] * x[i])
    
    return y_
```

```python
def error(X, y, theta):
    
    m = X.shape[0]
    err = 0.0
    
    for i in range(m):
        y_ = hypothesis(X[i], theta)
        err = err + (y[i] - y_)**2
    
    return err/m
```

```python
def gradient(X, y, theta):
    
    m, n = X.shape
    
    grad = np.zeros((n,))
    
    for j in range(n):
        for i in range(m):
        
            y_ = hypothesis(X[i], theta)
            grad[j]  = grad[j] + (y_ - y[i]) * X[i][j]
        
    return grad/m
```

```python
def gradientDescent(X, y, learning_rate = 0.1, max_epochs = 300):
    
    m, n = X.shape
    theta = np.zeros((n,))
    error_list = []
    
    for i in range(max_epochs):
        
        error_list.append(error(X, y, theta))
        grad_list = gradient(X, y, theta)
        
        for j in range(n):
            
            theta[j] = theta[j] - learning_rate*grad_list[j]
        
    return error_list, theta
        
```

```python
import time

start = time.time()
error_list, theta = gradientDescent(X, y)
end = time.time()

print("Time taken is: ", end - start)
```

```python
def r2_score(Y, Y_):
    
    num = np.sum((Y - Y_)**2)
    denom = np.sum((Y - Y.mean())**2)
    
    score = (1 - num/denom)
    return score*100
```

### Optimizing using Vectorization

```python
ones = np.ones((X.shape[0], 1))
X = np.hstack((ones, X))
```

```python
def hypothesis(x, theta):
    
    return np.dot(x, theta)
```

```python
def error(X, y, theta):
    
    e = 0.0
    m = X.shape[0]
    y_ = hypothesis(X, theta)
    e = np.sum((y - y_)**2)
    
    return e/m
```

```python
def gradient(X, y, theta):
    
    m = X.shape[0]
    y_ = hypothesis(X, theta)
    grad = np.dot(X.T, (y - y_))
    
    return grad/m
```

```python
def gradientDescent(X, y, learning_rate = 0.1, max_epochs = 300):
    
    n = X.shape[1]
    theta = np.zeros((n,))
    error_list = []
    
    for i in range(max_epochs):
        
        error_list.append(error(X, y, theta))
        grad_list = gradient(X, y, theta)
        theta = theta - learning_rate*grad_list
        
    return error_list, theta
        
```

```python
import time

start = time.time()
error_list, theta = gradientDescent(X, y)
end = time.time()

print('Total runtime is vectorized implementation : ', end - start)
```
